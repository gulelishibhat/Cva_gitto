```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
position = 0
walk = [position]
steps = 1000
for i in range (steps):
    step = 0.5 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

```python
nsteps = 1000
draws = np.random.randint(0,2,size=nsteps)
steps = np.where(draws > 0,1,-1)
walk = steps.cumsum()
```
